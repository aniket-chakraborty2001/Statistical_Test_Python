```python
# Importing Required Packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
# Reading the 'StatisticalTest_Data.csv' file

data = pd.read_csv('StatisticalTest_Data.csv')
data.head()
```

```python
# Checking the Data types of each column

data.dtypes
```

```python
# Removing the column 'Name' and 'Gender' from the data frame
# Storing the result in a data frame called data_num

cols = ['Name', 'Gender']
data_num = data.drop(cols, axis = 1)
print('Number of Columns in data_num:', data_num.shape[1])
data_num.head()
```

### Plotting the Correlation Plot - Called as Heatmap

* Calculate the Correlation Matrix
* Plot heatmap using sns library
* use the value of following parameters annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5

```python
# Calculate the Correlation Matrix using the .corr() method

cor_matrix = data_num.corr()
cor_matrix
```

```python
# Plotting the Correlation Plot using sns.heatmap function

plt.figure(figsize=(8,4))
sns.heatmap(cor_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Plot')
plt.show()
```
