```python
# Importing Required Packages

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
# Reading the data named as 'StatisticalTest_Data.csv'

data = pd.read_csv('StatisticalTest_Data.csv')
data.head()
```

```python
# Getting the Entries of the Individual Columns and Storing them into variables

math = data['Maths']
phy = data['Physics']
chem = data['Chemistry']
bio = data['Biology']
cs = data['ComputerScience']
ben = data['Bengali']
eng = data['English']
his = data['History']
geo = data['Geography']
stat = data['Statistics']
```

### Visualization for Bengali Data

```python
# Setting the Subplot Area to plot both plots
fig = plt.figure(figsize = (10,4))
ax0 = fig.add_subplot(1,2,1)
ax1 = fig.add_subplot(1,2,2)

#Boxplot for the Bengali Data
ben.plot(kind = 'box',ax = ax0)
ax0.set_title('Boxplot for Bengali Data')


# Histogram for Bengali data
ben.plot(kind = 'hist', bins = 15)
ax1.set_title('Histogram for Bengali Data')
ax1.set_xlabel('Bengali Score')

plt.show()
```
